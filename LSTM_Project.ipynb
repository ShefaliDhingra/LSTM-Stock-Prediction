## Problem Statement

Stock market prediction is a complex task due to the highly volatile and non-linear nature of financial data. Traditional statistical models struggle to capture sequential dependencies and long-term trends, leading to suboptimal forecasting accuracy.

This project aims to leverage **Long Short-Term Memory (LSTM)** networks, a specialized form of Recurrent Neural Networks (RNNs), to predict stock prices. By effectively modeling sequential dependencies and learning from historical data patterns, the LSTM model enhances forecasting accuracy.

The project focuses on predicting stock prices for **NIFTY 50 and HDFC Bank**, two key financial indicators in the Indian stock market. The implementation includes data preprocessing, LSTM model development, training, and evaluation to analyze the effectiveness of deep learning for stock price forecasting.

## Abstract

This project implements a **Long Short-Term Memory (LSTM)** model to predict stock prices based on historical market data. LSTM, a specialized type of Recurrent Neural Network (RNN), is well-suited for time-series forecasting due to its ability to retain long-term dependencies and mitigate the vanishing gradient problem.

The dataset includes stock prices for **NIFTY 50 and HDFC Bank**, obtained from Yahoo Finance (`yfinance`). The data is preprocessed, normalized, and transformed into a supervised learning format. The LSTM model is then trained on this data to learn patterns and make future price predictions.

Model performance is evaluated using **Mean Squared Error (MSE) and Mean Absolute Error (MAE)**, with predictions visualized against actual stock prices. The results highlight the potential of deep learning models in financial forecasting, demonstrating their ability to capture trends and fluctuations in stock prices.

This study provides insights into the practical applications of LSTM-based models in financial markets, showcasing their ability to improve decision-making for traders, investors, and financial analysts.

```python
import numpy as np
import datetime
import pandas as pd

def hb11_str_to_datetime(hb11_s):
    hb11_split = hb11_s.split('-')
    hb11_year, hb11_month, hb11_day = int(hb11_split[0]), int(hb11_split[1]), int(hb11_split[2])
    return datetime.datetime(year=hb11_year, month=hb11_month, day=hb11_day)

def hb11_df_to_windowed_df(hb11_dataframe, hb11_first_date_str, hb11_last_date_str, hb11_n=3):
    hb11_first_date = hb11_str_to_datetime(hb11_first_date_str)
    hb11_last_date = hb11_str_to_datetime(hb11_last_date_str)

    hb11_dates = []
    hb11_X, hb11_Y = [], []

    hb11_current_date = hb11_first_date

    while hb11_current_date <= hb11_last_date:
        hb11_window = []
        hb11_temp_date = hb11_current_date

        # Find the values for the previous n valid trading days
        for i in range(hb11_n):
            hb11_temp_date -= datetime.timedelta(days=1)
            while hb11_temp_date >= hb11_dataframe.index[0]:
                hb11_temp_date_dt = pd.to_datetime(hb11_temp_date)
                if hb11_temp_date_dt in pd.to_datetime(hb11_dataframe.index):
                    hb11_window.insert(0, hb11_dataframe.loc[hb11_temp_date_dt])
                    break
                hb11_temp_date -= datetime.timedelta(days=1)
            else: #break out if we are out of dates.
                break

        # Check if we have enough data for the window
        if len(hb11_window) == hb11_n and pd.to_datetime(hb11_current_date) in pd.to_datetime(hb11_dataframe.index):
            hb11_X.append(hb11_window)
            hb11_Y.append(hb11_dataframe.loc[pd.to_datetime(hb11_current_date)])
            hb11_dates.append(hb11_current_date)

        hb11_current_date += datetime.timedelta(days=1)

    hb11_ret_df = pd.DataFrame({})
    hb11_ret_df['Target Date'] = hb11_dates

    hb11_X = np.array(hb11_X)
    for hb11_i in range(0, hb11_n):
        hb11_ret_df[f'Target-{hb11_n-hb11_i}'] = hb11_X[:, hb11_i]

    hb11_ret_df['Target'] = hb11_Y

    return hb11_ret_df
```

```python
import numpy as np

def hb11_windowed_df_to_date_X_y(hb11_windowed_dataframe):
    hb11_df_as_np = hb11_windowed_dataframe.to_numpy()

    hb11_dates = hb11_df_as_np[:, 0]

    hb11_middle_matrix = hb11_df_as_np[:, 1:-1]
    hb11_X = hb11_middle_matrix.reshape((len(hb11_dates), hb11_middle_matrix.shape[1], 1))

    hb11_Y = hb11_df_as_np[:, -1]

    return hb11_dates, hb11_X.astype(np.float32), hb11_Y.astype(np.float32)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a scaler object for Nifty
hb11_scaler_nifty = MinMaxScaler()

# Reshape the data for scaling (if necessary)
hb11_X_train_nifty_reshaped = hb11_X_train_nifty.reshape(-1, 1)
hb11_y_train_nifty_reshaped = hb11_y_train_nifty.reshape(-1, 1)

# Fit the scaler to the training data and transform it
hb11_X_train_nifty_scaled = hb11_scaler_nifty.fit_transform(hb11_X_train_nifty_reshaped).reshape(hb11_X_train_nifty.shape)
hb11_y_train_nifty_scaled = hb11_scaler_nifty.fit_transform(hb11_y_train_nifty_reshaped).reshape(hb11_y_train_nifty.shape)


# Scale validation and testing data using the fitted scaler
hb11_X_val_nifty_scaled = hb11_scaler_nifty.transform(hb11_X_val_nifty.reshape(-1, 1)).reshape(hb11_X_val_nifty.shape)
hb11_y_val_nifty_scaled = hb11_scaler_nifty.transform(hb11_y_val_nifty.reshape(-1, 1)).reshape(hb11_y_val_nifty.shape)

hb11_X_test_nifty_scaled = hb11_scaler_nifty.transform(hb11_X_test_nifty.reshape(-1, 1)).reshape(hb11_X_test_nifty.shape)
hb11_y_test_nifty_scaled = hb11_scaler_nifty.transform(hb11_y_test_nifty.reshape(-1, 1)).reshape(hb11_y_test_nifty.shape)
```

```python
# Model for Nifty
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras.callbacks import ReduceLROnPlateau

hb11_model_nifty = Sequential([
    layers.Input((3, 1)),
    layers.LSTM(128, return_sequences=True),  # Increased units, return sequences for stacked LSTM
    layers.LSTM(64),  # Added another LSTM layer
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),  # Added dropout for regularization
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

# Learning rate scheduler
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)

hb11_model_nifty.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

hb11_model_nifty.fit(hb11_X_train_nifty_scaled, hb11_y_train_nifty_scaled,
                    validation_data=(hb11_X_val_nifty_scaled, hb11_y_val_nifty_scaled),
                    epochs=100,
                    callbacks=[lr_scheduler])
```

```python
# Model for HDFC
hb11_model_hdfc = Sequential([layers.Input((3, 1)),
                          layers.LSTM(64),
                          layers.Dense(32, activation='relu'),
                          layers.Dense(32, activation='relu'),
                          layers.Dense(1)])

hb11_model_hdfc.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

hb11_model_hdfc.fit(hb11_X_train_hdfc, hb11_y_train_hdfc, validation_data=(hb11_X_val_hdfc, hb11_y_val_hdfc), epochs=100)
```
